```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
from sklearn.model_selection import train_test_split
```

```python
# Loading and merging the dataset might take few mintues
Songs_DataFrame1 = pd.read_table("10000.txt",header=None)
Songs_DataFrame1.columns = ['user_id', 'song_id', 'listen_count']

#pandas.read function used to read the given data
Songs_DataFrame2 =  pd.read_csv("song_data.csv")
```

```python
# To create input DataFrame , the above two DataFrames need to be combined
Songs_DataFrame = pd.merge(Songs_DataFrame1, Songs_DataFrame2.drop_duplicates(['song_id']), on="song_id", how="left")
```

```python
# To form a merged column , title of the song and its perticular  artist name should be merged
Songs_DataFrame['song'] = Songs_DataFrame['title'].map(str) + " - " + Songs_DataFrame['artist_name']
```

# **Data Cleaning**

```python
Songs_DataFrame.isna()
```

```python
Songs_DataFrame.info()
```

```python
# Handle missing values by dropping rows with null user_id or song_id
Songs_DF = Songs_DataFrame.dropna()
# To form a merged column , title of the song and its perticular  artist name should be merged
Songs_DF['song'] = Songs_DF['title'].map(str) + " - " + Songs_DF['artist_name']
display(Songs_DF)
```

```python
# Select the relevant columns and rename them
Songs = Songs_DF[['user_id', 'song_id', 'listen_count']]
Songs.columns = ['user_id', 'song_id', 'rating']
```

```python
# Ensure listen_count is numeric
Songs['rating'] = pd.to_numeric(Songs['rating'], errors='coerce')

# Drop rows with invalid listen_count values
df = Songs.dropna(subset=['rating'])
```

```python
from sklearn.preprocessing import MinMaxScaler
# Step 1: Normalization
scaler = MinMaxScaler()
Songs['normalized_rating'] = scaler.fit_transform(df[['rating']])

# Step 2: Filtering
# Define thresholds for minimum interactions
min_user_interactions = 5
min_song_interactions = 5

# Filter users
user_interaction_counts = Songs['user_id'].value_counts()
users_to_keep = user_interaction_counts[user_interaction_counts >= min_user_interactions].index
Songs = Songs[Songs['user_id'].isin(users_to_keep)]

# Filter songs
song_interaction_counts = Songs['song_id'].value_counts()
songs_to_keep = song_interaction_counts[song_interaction_counts >= min_song_interactions].index
Songs = Songs[Songs['song_id'].isin(songs_to_keep)]

# Step 3: Handling Sparsity
# Ensure the dataset is not too sparse
min_interactions = 5

# Refilter users
user_interaction_counts = Songs['user_id'].value_counts()
users_to_keep = user_interaction_counts[user_interaction_counts >= min_interactions].index
Songs = Songs[Songs['user_id'].isin(users_to_keep)]

# Refilter songs
song_interaction_counts = Songs['song_id'].value_counts()
songs_to_keep = song_interaction_counts[song_interaction_counts >= min_interactions].index
Songs = Songs[Songs['song_id'].isin(songs_to_keep)]
```

# **Model Training using SVD**

```python
from sklearn.model_selection import train_test_split
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import svds

# Créer un échantillon de données
sample_size = 10000  # Définir la taille de l'échantillon, ajuster selon vos besoins
Songs = Songs.sample(n=sample_size, random_state=42)

# Create a user-item interaction matrix
user_item_matrix = Songs.pivot(index='user_id', columns='song_id', values='normalized_rating').fillna(0)

# Split the data into training and testing sets
train_data, test_data = train_test_split(Songs, test_size=0.2, random_state=42)

# Convert the user-item matrix to a sparse matrix format
user_item_matrix_sparse = csr_matrix(user_item_matrix.values)

# Perform SVD
U, sigma, Vt = svds(user_item_matrix_sparse, k=70)

# Convert sigma to a diagonal matrix
sigma = np.diag(sigma)

# Calculate predicted ratings
predicted_ratings = np.dot(np.dot(U, sigma), Vt)

# Create a DataFrame for predicted ratings
predicted_ratings_df = pd.DataFrame(predicted_ratings, columns=user_item_matrix.columns, index=user_item_matrix.index)
```

```python
# Define the recommendation function
def recommend_songs(user_id, predicted_ratings_df, original_ratings_df, original_df, num_recommendations=10):
    user_row_number = user_id
    sorted_user_predictions = predicted_ratings_df.loc[user_row_number].sort_values(ascending=False)

    user_data = original_ratings_df[original_ratings_df.user_id == user_id]
    user_full = user_data.merge(pd.DataFrame(sorted_user_predictions).reset_index(), how='left', on='song_id')
    user_full = user_full.rename(columns={user_row_number: 'Predictions'}).sort_values('Predictions', ascending=False)

    recommendations = sorted_user_predictions[~sorted_user_predictions.index.isin(user_full['song_id'])].sort_values(ascending=False).head(num_recommendations)

    # Merge with the original dataset to get song names
    recommendations = recommendations.reset_index()
    recommendations = recommendations.merge(Songs_DF[['song_id', 'song']], on='song_id', how='left').drop_duplicates('song_id')

    return user_full, recommendations

# Get recommendations for a specific user
user_full, recommendations = recommend_songs('59d4bcae89ffb657c97c69d69d931a459136489a', predicted_ratings_df, Songs, 10)

#print("User's Data:")
#print(user_full)
print("\nTop 10 Recommendations:")
print(recommendations['song'])
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

def rmse(predicted_ratings, true_ratings):
    predicted_ratings = predicted_ratings[true_ratings.nonzero()].flatten()
    true_ratings = true_ratings[true_ratings.nonzero()].flatten()
    return sqrt(mean_squared_error(predicted_ratings, true_ratings))

# Get true ratings for the test set
test_user_item_matrix = test_data.pivot(index='user_id', columns='song_id', values='rating').fillna(0)
test_user_item_matrix_sparse = csr_matrix(test_user_item_matrix.values)

# Compute RMSE
test_rmse = rmse(predicted_ratings_df.values, test_user_item_matrix_sparse.toarray())
print(f'Test RMSE: {test_rmse}')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import TruncatedSVD
import gc

# Supposons que Songs_DF est déjà chargé avec les colonnes 'song_id', 'song', 'title', 'artist_name'
# Supposons que predicted_ratings_df et original_ratings_df sont déjà chargés

# Vectoriser les caractéristiques combinées
vectorizer = CountVectorizer()
song_feature_matrix = vectorizer.fit_transform(Songs_DF['song'])

# Réduire les dimensions avec TruncatedSVD
svd = TruncatedSVD(n_components=100)
song_feature_matrix_reduced = svd.fit_transform(song_feature_matrix)

# Calculer la similarité cosinus entre les chansons
song_similarity = cosine_similarity(song_feature_matrix_reduced)

# Définir la fonction de recommandation
def recommend_songs(user_id, predicted_ratings_df, original_ratings_df, original_df, song_similarity, num_recommendations=10):
    user_row_number = user_id
    sorted_user_predictions = predicted_ratings_df.loc[user_row_number].sort_values(ascending=False)

    user_data = original_ratings_df[original_ratings_df.user_id == user_id]
    user_full = user_data.merge(pd.DataFrame(sorted_user_predictions).reset_index(), how='left', on='song_id')
    user_full = user_full.rename(columns={user_row_number: 'Predictions'}).sort_values('Predictions', ascending=False)

    recommendations = sorted_user_predictions[~sorted_user_predictions.index.isin(user_full['song_id'])].sort_values(ascending=False).head(num_recommendations)

    # Fusionner avec le dataset original pour obtenir les noms de chansons
    recommendations = recommendations.reset_index()
    recommendations = recommendations.merge(original_df[['song_id', 'title', 'artist_name']], on='song_id', how='left').drop_duplicates('song_id')

    # Incorporer le filtrage basé sur le contenu en trouvant des chansons similaires à celles déjà appréciées par l'utilisateur
    user_songs = user_full['song_id'].tolist()
    similar_songs = []
    for song in user_songs:
        similar_songs.extend(song_similarity[Songs_DF[Songs_DF['song_id'] == song].index[0]].argsort()[-(num_recommendations+1):-1])

    similar_songs = list(set(similar_songs) - set(user_songs))[:num_recommendations]
    similar_songs = Songs_DF.iloc[similar_songs]['song_id'].tolist()

    recommendations['content_based'] = similar_songs

    return user_full, recommendations

# Obtenir des recommandations pour un utilisateur spécifique
user_full, recommendations = recommend_songs('59d4bcae89ffb657c97c69d69d931a459136489a', predicted_ratings_df, Songs[['user_id', 'song_id', 'normalized_listen_count']], Songs_DF, song_similarity, 10)

print("User's Data:")
print(user_full)
print("\nTop 10 Recommendations:")
print(recommendations[['song_id', 'title', 'artist_name', 'content_based']])

# Libérer de la mémoire inutilisée
del song_feature_matrix
gc.collect()
```
